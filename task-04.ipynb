```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
```

```python
df = pd.read_csv("/content/adult.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
categorical_col = df.select_dtypes(include=["object"]).columns
numerical_col = df.select_dtypes(include=["int64", "float64"]).columns
```

```python
print("Categorical Columns: ", categorical_col)
print("Numerical Columns: ", numerical_col)
```

```python
from sklearn.preprocessing import LabelEncoder

Le = LabelEncoder()
df["income"] = Le.fit_transform(df["income"])
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_col, drop_first=True)

print(df_encoded.head())
```

```python
scaler = StandardScaler()
df_encoded[numerical_col] = scaler.fit_transform(df_encoded[numerical_col])
```

```python
#befor Scaling
df[numerical_col].describe()
```

```python
#After Scaling
df_encoded[numerical_col].describe()
```

```python
df_encoded.to_csv("adult_preprocessed.csv", index=False)
print("Processed dataset saved successfully!")
```
